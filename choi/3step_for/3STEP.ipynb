#2739문제

```python
N = int(input())
for i in range(1,10):
    print(N,'*',i,'=',N*i)
```

```python
# 구구단 정답 앞 한칸 비게 하고 싶을떄
N = int(input())
for i in range(1,10):
    print("%d * %d = %2d"%(N,i,N*i))
```

#2742문제

```python
N = int(input())
sum = N
for i in range(0,N+1):
    print(sum)
    sum -= 1
    if sum == 0:
        break
```

#10871문제

```python
N, x = map(int,input().split())
num = list(map(int,input().split()))
for i in range(N):
    if num[i] < x:
        print(num[i],end=' ')
```
